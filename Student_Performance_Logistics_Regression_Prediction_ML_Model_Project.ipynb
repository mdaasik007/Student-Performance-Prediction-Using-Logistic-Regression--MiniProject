```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

data = pd.read_csv("student_performance_pass_fail.csv")
le = LabelEncoder() #Convert values to numerical values
data["Passed"] = le.fit_transform(data["Passed"])

X = data[["Hours_Studied","Attendance_Percentage","Previous_Score"]]
y = data["Passed"]

model = LogisticRegression()
model.fit(X,y)

# Test Area
new_student = pd.DataFrame({
    "Hours_Studied":[4],
    "Attendance_Percentage":[73],
    "Previous_Score":[77]
})


#Prediction
prediction = model.predict(new_student)
print("Encoded Prediction:", prediction)
print("Decode Prediction:", le.inverse_transform(prediction))
print("Probablity:", model.predict_proba(new_student))
```
